```python
#判断是否满足子集的条件,
#找出所有子集
def inn(ls,items):
    N = len(items)
    flag = 0
    for i in range(2 ** N):
        combo = []
        for j in range(N):
            if (i >> j) % 2 :
                combo.append(items[j])
#         print(combo)
        if ls == combo:
            flag += 1
        else:
            flag += 0
    if flag > 0:
        return True
    else:
        return False
```

```python
#s所有子集列表
s=[]
for q in data:
    N = len(q)
    for i in range(2**N):
        combo = []
        for j in range(N):
            if (i>>j)%2:
                combo.append(q[j])
                s.append(combo)
s
```

```python
#列表去重之后，原始数据集中的所有子集
com=[]
for i in s:
    if i not in com:
        com.append(i)
com
```

```python
#计算子集出现的频数
def scan(data,com): 
    dic={}
    for i in com:
        for j in data:
            if inn(i,j):
                dic[str(i)] = dic.get(str(i),0) + 1
    return dic
```

```python
#产生庞大的数据集，把没有出现过的数据集删除
def get_support(data,min_support):
    dic = {}
    dic.update(scan(data,com))
#     dic
    num = len(data)    
    support_dic = {}
    count = 0
    for key,value in dic.items():
        support = value/num
        if support >= min_support:
            support_dic[key] =support
            count += 1 
    return support_dic
```

```python
def get_rlue(support_dic):
    count = 0
    for key,value in support_dic.items():#支持度频繁项集
        n = len(eval(key))
        if  n == 2 :
            front = eval(key)[0]
            after = eval(key)[1]
            confi = support_dic[key]/support_dic[str([front])]
            print(f'{[front]}-->{[after]}:{confi}')
        elif n > 2:
            front = eval(key)[:n-1]
            after = eval(key)[-1]
            confi = support_dic[key]/support_dic[str(front)]
            print(f'{front}-->{[after]}:{confi}')
        count += 1
```
