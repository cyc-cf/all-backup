```python
#sigmoid函数
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

#例1
X = np.linspace(-5,5,200)             #linsapce固定范围内的等间隔的数
y = [1/(1+math.e**(-x)) for x in X]
plt.plot(X,y)
plt.show()

#例2
X = np.linspace(-60,60,200)
y = [1/(1+math.e**(-x)) for x in X]
plt.plot(X,y)
plt.show()
```

```python
逻辑回归=线性模型+sigmoid函数
```

```python
# Y=1 损失函数
X = np.linspace(0.0001,1,200)
y = [(-np.log(x)) for x in X]
plt.plot(X,y)
plt.show()

# Y=0 损失函数
X = np.linspace(0,0.99999,200)
y = [(-np.log(1-x)) for x in X]
plt.plot(X,y)
plt.show()
```
