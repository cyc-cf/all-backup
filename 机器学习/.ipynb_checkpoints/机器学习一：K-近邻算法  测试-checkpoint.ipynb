K近邻算法：
给定一个标签类别的训练数据集，输入没有标签的新数据，
在训练数据集中找到与新数据最邻近的k个实例，在数据集中找到与新数据最邻近
的K个实例，若这K个实例的多数属于某个类别，那么新数据就属于这个类别。

### 3 数据归一化

```python
#数据归一化
def minmax(dataSet):
    minDf = dataSet.min()
    maxDf = dataSet.max()
    normSet = (dataSet - minDf)/(maxDf - minDf)
    return normSet
```

```python
datingT = pd.concat([minmax(datingTest.iloc[:,:3]),datingTest.iloc[:,3]],axis=1)
datingT.head()
```

### 4 划分训练集和测试集

```python
训练集：训练模型
测试集：验证模型的准确性
```

```python
datingT
```

```python
'''
函数功能：切分训练集和测试集
rate:训练集所占的比例
'''
def randsplit(dataSet,rate=0.9):
    n = dataSet.shape[0]   #总数据
    m = int(n*rate)        #训练集数据
    train = dataSet.iloc[:m,:n]  #取m行数据
    test = dataSet.iloc[m:,:]    #取n-m行数据
    test.index = range(test.shape[0])  #对n-m行数据重新计数
    return train,test
```

```python
train,test = randsplit(datingT)
train
```

## 三、约会网站配对效果判定

```python
def datingClass(train,test,k):
    n = train.shape[1]-1  #3
    m = test.shape[0]   #100
    result =[]
    for i in range(m):
        dist = list((((train.iloc[:,:n] - test.iloc[i,:n]) ** 2).sum(1))**0.5)
        dist_1 = pd.DataFrame({'dist':dist,'labels':(train.iloc[:,n])})
        dr = dist_1.sort_values(by ='dist')[:k]
        re = dr.loc[:,'labels'].value_counts()
        result.append(re.index[0])
    result = pd.Series(result)
    test['predict'] = result
    acc = (test.iloc[:,-1]== test.iloc[:,-2]).mean()
    print(f'模型预测准确率未{acc}')
    return test
```

```python
datingClass(train,test,5)
```
