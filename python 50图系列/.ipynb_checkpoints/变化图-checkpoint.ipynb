```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Simhei'] 
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/economics.csv")
df.head()
```

### 副座标轴的折线图

```python
x = df['date'] #横坐标
y1 = df['psavert']
y2 = df['unemploy']
```

```python
fig,ax1 =plt.subplots(1,1,figsize=(16,9),dpi=80)#一行一列只生成一个画布和一个子图
ax1.plot(x,y1,color='tab:red')

ax2 = ax1.twinx() #实例化共享相同x轴的第二个轴
ax2.plot(x,y2,color='tab:gray')
#装饰图像
ax1.set_xlabel('年份',fontsize=12)
ax1.tick_params(axis='x',rotation=0,labelsize=12)#设置与标尺有关的参数
ax1.set_ylabel('数据',color='tab:red',fontsize=12)
ax1.tick_params(axis='y',rotation=0,
                labelcolor='tab:red')#调节坐标轴标尺的颜色
ax1.grid(alpha=0.4)

ax2.set_ylabel('人数', color='tab:gray',fontsize=12)
ax2.tick_params(axis='y',labelcolor='tab:gray')#坐标轴标尺的颜色

ax2.set_xticks(np.arange(0,len(x),30)) #x坐标轴刻度
ax2.set_xticklabels(x[::30], rotation=90, fontdict={'fontsize':10})
#ax2.set_title("人数 vs 数据",fontsize=22)
fig.tight_layout()
plt.show()
```

### 不堆积的面积图

```python
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/economics.csv")
```

```python
x = df['date'].values.tolist()
y1 = df['psavert'].values.tolist()
y2 = df['uempmed'].values.tolist()
```

```python
mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive']      
columns = ['个人储蓄率', '失业时长中位数']
#填充阴影
plt.fill_between(x,y1=y1,y2=0, #定义三个坐标
                 label=columns[0],
                 alpha=0.5,
                 color=mycolors[0],
                 linewidth=2)
#绘制横线
for y in np.arange(2.5, 30.0, 2.5): #一次只能绘制一条线，要循环绘制
    plt.hlines(y,xmin=0,xmax=len(x),colors='black',alpha=0.3,linestyles='--',lw=0.5)
;
```
