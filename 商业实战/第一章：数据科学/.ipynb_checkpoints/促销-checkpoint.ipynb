```python
import pandas as pd
import numpy as np
trad_flow = pd.read_csv(r'E:\Ajiao\LiXian\资料\技术实详解与商业实战\280_Ben_八大直播八大案例配套课件\第一章\RFM_TRAD_FLOW.csv',encoding = 'gbk')
trad_flow.head(20)
# 包含客户  交易金额  交易类型  交易时间点
```

```python
trad_flow['type_label'].value_counts()
```

```python
# 看客户兴趣：看打折的产品占整个购买产品的比例
F = trad_flow.groupby(['cumid','type']).agg({'transID':'count'})
F
```

```python
# 纵向数据转为横向数据，方便计算
```

```python
F_trans=pd.pivot_table(F,index='cumid',columns='type',values='transID')
F_trans.head()
```

```python
# 缺失值替换:未购买特价商品记为0
F_trans['Special_offer']= F_trans['Special_offer'].fillna(0)
F_trans.head()
```

```python
# 兴趣度：特价商品/（特价商品+正常商品）
F_trans["interest"]=F_trans['Special_offer']/(F_trans['Special_offer']+F_trans['Normal'])
F_trans.head()
```

```python
# 价值信息：购买不同产品的金额
money = trad_flow.groupby(['cumid','type']).agg({'amount':'sum'})
money
```

```python
data = pd.pivot_table(money,index='cumid',columns='type',values='amount')
data
```

```python
data['Special_offer']=data['Special_offer'].fillna(0)
data['returned_goods']=data['returned_goods'].fillna(0)
data
```

```python
data['value'] = data['Special_offer']/(data['Normal']+data['returned_goods'])
data
```

```python
trad_flow
```

```python
# 计算每个客户是否是活跃的：最后一次消费记录日期是否久远
from datetime import datetime
import time
def to_time(t):
    out_t = time.mktime(time.strptime(t,'%d%b%y:%H:%M:%S'))
    return out_t
```

```python
trad_flow["time_new"]= trad_flow.time.apply(to_time)
trad_flow.head()
```

```python
R = trad_flow.groupby(['cumid']).agg({'time_new':'max'})
R
```

```python
# 构建模型：筛选目标用户
from sklearn import preprocessing
threshold = pd.qcut(F_trans['interest'],2,retbins = True)[1][1] # 分箱操作
binarizer = preprocessing.Binarizer(threshold = threshold)   #进行二分类
interest_q = pd.DataFrame(binarizer.transform(F_trans['interest'].values.reshape(-1,1)))# 二值化
interest_q.index = F_trans.index
interest_q.columns=['interest']
```

```python
threshold = pd.qcut(data['value'], 2, retbins=True)[1][1]
binarizer = preprocessing.Binarizer(threshold=threshold)
value_q = pd.DataFrame(binarizer.transform(data['value'].values.reshape(-1, 1)))
value_q.index=data.index
value_q.columns=["value"]
```

```python
threshold = pd.qcut(R["time_new"], 2, retbins=True)[1][1]
binarizer = preprocessing.Binarizer(threshold=threshold)
time_new_q = pd.DataFrame(binarizer.transform(R["time_new"].values.reshape(-1, 1)))
time_new_q.index=R.index
time_new_q.columns=["time"]
```

```python
analysis = pd.concat([interest_q,value_q,time_new_q],axis=1)
analysis
```

```python
analysis['rank']=analysis.interest+analysis.interest
analysis
```

```python
label = {
    (0,0,0):'无兴趣-低价值-沉默',
    (1,0,0):'有兴趣-低价值-沉默',
    (1,0,1):'有兴趣-低价值-活跃',
    (0,0,1):'无兴趣-低价值-活跃',
    (0,1,0):'无兴趣-高价值-沉默',
    (1,1,0):'有兴趣-高价值-沉默',
    (1,1,1):'有兴趣-高价值-活跃',
    (0,1,1):'无兴趣-高价值-活跃'
}
analysis['label'] = analysis[['interest','value','time']].apply(lambda x: label[(x[0],x[1],x[2])], axis = 1)
analysis.head()
```
