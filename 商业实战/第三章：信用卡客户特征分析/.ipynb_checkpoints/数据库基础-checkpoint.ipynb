```python
import pandas as pd
import os 
import numpy as np
os.chdir(r"E:\Ajiao\LiXian\资料\技术实详解与商业实战\280_Ben_八大直播八大案例配套课件\第三章\提交-第三讲：信用卡客户特征分析-产品客户画像初步\提交-第三讲：信用卡客户特征分析-产品客户画像初步\数据与脚本\5Preprocessing")
camp = pd.read_csv('teleco_camp_orig.csv')
camp
```

```python
camp.describe()
```

```python
vmean = camp['Age'].mean(axis=0, skipna=True)# 提取均值，skipna=True计算去掉缺失值的均值
camp['Age_empflag'] = camp['Age'].isnull()#产生虚拟指标，判断是否补的缺失值
camp['Age']= camp['Age'].fillna(vmean)# 用均值填补缺失值
camp['Age'].describe()
```

```python
处理缺失值的原则就是：补缺失值+虚拟指标    然后建模
```
