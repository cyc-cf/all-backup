```python
# 客户收入是否对开卡有影响
import os
import pandas as pd

os.chdir(r'E:\Ajiao\LiXian\资料\技术实详解与商业实战\280_Ben_八大直播八大案例配套课件\第四章\提交-第四讲：统计建模与分析报告-二手房价格分析报告\脚本\6inference')
creditcard = pd.read_csv(r'creditcard_exp.csv',encoding='gbk')
creditcard
```

```python
两样本T检验：看一个分类变量给另一个分类变量(或一个连续变量) 的均值是否有差异
```

```python
#收入描述统计
creditcard['Income'].groupby(creditcard['Acc']).describe()
```

```python
# 两样本T检验
#第一步：方差齐次检验
from scipy import stats
Suc0 = creditcard[creditcard['Acc'] == 0]['Income']  #未开卡的人的收入
Suc1 = creditcard[creditcard['Acc'] == 1]['Income']  #开卡人的收入
leveneTestRes = stats.levene(Suc0,Suc1,center='median')
print('a-value=%6.4f,p-value=%6.4f'%leveneTestRes)
```

```python
#进行T检验：结果中statistic指的是T值， pvalue指的是p值
stats.stats.ttest_ind(Suc0, Suc1, equal_var=False)
```

```python
p值很小，所以
第一步：原假设是开卡人的平均收入=未开卡人的平均收入
第二部：显著度：样本是150 所以显著度设为5%
第三步：样本采集
第四步：计算统计量 t = 【（样本量成功-不成功）-（开卡人平均收入-不开卡人平均收入）】/标准误
        也就是第一个样本均值-第二个样本均值-0（总体差值）/样本均值相减的标准误
```
