```python
import jieba
import jieba.analyse
import jieba.posseg
```

```python
#1 去停用词和虚词
def stopwordslist():
    stopwordslist = [line.strip() for line in open('E:\Ajiao\LiXian\学长任务\文本挖掘之隐患数据库\虚词和停用词.txt',encoding='UTF-8').readlines()]
    return stopwordslist
```

```python
def seg_depart(sen):
    outstr = ''
    if isinstance(sen,str):
        sen_depart = jieba.cut(sen.strip())
        stopwords = stopwordslist()
        for word in sen_depart:
            if word not in stopwords:
                if word != '\t':
                    outstr += word
                    outstr +=''
    return outstr
```

```python
outfilename = 'E:\Ajiao\LiXian\学长任务\文本挖掘之隐患数据库\out.txt'
outputs = open(outfilename,'w',encoding = 'UTF-8')
for i in [data['问题描述'].iloc[i] for i in range(len(data)) if i <= len(data)]:
    empty_word = seg_depart(i)
    outputs.write(empty_word+'\n')
outputs.close()
print('删除虚词和停用词成功！！！')
```

```python
# 2 读文件
f = open('E:\Ajiao\LiXian\学长任务\文本挖掘之隐患数据库\out.txt','r',encoding='UTF-8')
file = f.read()
print(file)
```

```python
# 4.去除时间和英文字母和换行符和年月日
result = re.sub(r'(\d+|[a-zA-Z]+|\n+|\u5e74+|\u6708+|\u65e5)','',file)
result
```

```python
# 5 分词
data_lcut=jieba.lcut(result,cut_all=False)
data_lcut
```

```python
# 6 专业术语词库
dict_1= 'E:\Ajiao\LiXian\学长任务\文本挖掘之隐患数据库\专业术语词汇.txt'
jieba.load_userdict(dict_1)
f = open(dict_1,'r',encoding='UTF-8')
for i in f.readlines():
    jieba.suggest_freq(i,tune=True)
```

```python
# 提取特征项
r = jieba.analyse.extract_tags(result,topK=50,allowPOS=('x','p','r','u','c','e','a','d','f','g'))
trait='/'.join(r)
trait
```

```python
# 词频
cou = []
word_list=jieba.lcut(result)
for i in range(len(r)):
    if r[i] in word_list:
        cou.append(word_list.count(r[i]))
        i=i+1
cou
```

```python
data_trai=pd.DataFrame({'频数':cou},index=r)
da_trait=data_trai.sort_values(by=['频数'],ascending=False)
da_trait
```
